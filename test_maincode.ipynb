### プロジェクトごとに git log を取る手法

```python
## 変更点に関してコメントが含まれるかどうか判定するモジュール
## 最初の段階では、全てのコメントが一度にそのまま追加されると想定する
def CheckKeyWords(log, comments):
    for comment in comments:
        if comment in log:
            pass
        else:
            return False
    return True
```

```python
startRevision = "664a7f4730f58be00fd7fe0526d2c13180107c6f" 
comments = "# Add a user with an explicit UID/GID and create necessary directories"
comments = comments.splitlines()

cwd = './tmp/' + project.split('/')[1]
cmd = f'git log -p'

logs = subprocess.run(list(cmd.split()), cwd=cwd, encoding='utf-8', stdout=subprocess.PIPE, errors="ignore")
# logs = logs.stdout.split('\ncommit ')
logs = re.split('\ncommit\s|\n\s{4}commit\s', logs.stdout)
# print(logs)

## 指定のリビジョンを発見すると開始する
startFLG = False
finishFLG = False

results = {"commitid":[], "date":[], "file":[], "diff":[], "rename": []}

for num, log in enumerate(logs):
    if num == 0:
        commitid = log[7:47]
    else:
        commitid = log[0:40]
    
    if commitid == startRevision:
        startFLG = True
    if startFLG:
        date = re.findall(r'Date:\s*([^\n]*)\n', log)

        results["commitid"].append(commitid)
        results["date"].append(date)

        discover = CheckKeyWords(log, comments)
        
        changefile = re.findall(r'diff\s--git\s([^\n]*)\n', log)
        renamefile = re.findall(r'rename\sfrom\s([^\n]*)\nrename\sto\s([^\n]*)\n', log)
        results["rename"].append(renamefile)
        
        results["file"].append(changefile)

        if discover:
            results["diff"].append(log)
        else:
            results["diff"].append(None)
        
    else:
        pass

print(results)
```

### 対象ファイルごとに git log を取る手法
コピーなどの要因もあり、きちんとしたログが出ない可能性があるため却下

```python
## 変更点に関してコメントが含まれるかどうか判定するモジュール
## 最初の段階では、全てのコメントが一度にそのまま追加されると想定する
def CheckKeyWords(log, comments):
    for comment in comments:
        if comment in log:
            pass
        else:
            return False
    return True
```

```python
projectpaths = [
#     "31z4/storm-docker/1.1.3/Dockerfile",
#     "31z4/storm-docker/1.2.3/Dockerfile",
#     "31z4/storm-docker/2.1.0/Dockerfile"
]

startRevision = "664a7f4730f58be00fd7fe0526d2c13180107c6f" 
comments = "# Add a user with an explicit UID/GID and create necessary directories"
comments = comments.splitlines()


for projectpath in projectpaths:
    print('\n-------- start --------')
    print(projectpath)
    
    ## 対象ファイルに git log
    ## ./ファイル名指定. (プロジェクト名は削除)
    projectpath = projectpath[len(project):]
    cwd = "./tmp/" + project.split('/')[1]
    cmd = f'git log -p .{projectpath}'
    logs = subprocess.run(list(cmd.split()), cwd=cwd, encoding='utf-8', stdout=subprocess.PIPE)
    logs = logs.stdout.split('commit')

    ## 指定のリビジョンを発見すると開始する
    startFLG = False
    finishFLG = False

    result = {"commitid":[], "date":[]}

    for log in logs[1:]:
        commitid = log[1:41]
        if commitid == startRevision:
            startFLG = True
        if startFLG:
            date = re.findall(r'Date:\s*([^\n]*)\n', log)[0]

            result["commitid"].append(commitid)
            result["date"].append(date)


            discover = CheckKeyWords(log, comments)

            if discover: ##追加された日付
                finishFLG = True
                print('============   FINISH   ================')
                break
            else:
                pass
        else:
            pass


    if not finishFLG:
        print('>>>>>>>>> Not Finished >>>>>>>>>>>>')

    print(result)
```
